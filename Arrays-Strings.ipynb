1. Merge Strings Alternately (EASY)

```python
# First solution
# 20 MS, beats 30%, JOINING AT END IS HEAVY

class Solution(object):
    def mergeAlternately1(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: str
        """
        s = []
        max_length = max(len(word1), len(word2))

        for letter in range(max_length):
            if letter < len(word1):
                s.append(word1[letter])
            if letter < len(word2):
                s.append(word2[letter])
        return "".join(s)

# Using two pointers 
# 16 MS, beats 30%
class Solution:
    def mergeAlternately2(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: str
        """
        result = ""
        i = 0 
        j = 0

        while i < len(word1) or j < len(word2):
            if  i < len(word1):
                result += word1[i]
                i += 1
            if j < len(word2):
                result += word2[j]
                j+=1
        return "".join(result)
    
# Using while loop and slicing 
# 7ms, beats 90%
class Solution:
    def mergeAlternately(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: str
        """
        result = ""
        i = j = 0
        while i < len(word1) or j < len(word2):
            result += word1[i:i+1] + word2[j:j+1]
            i += 1
            j += 1
        while i < len(word1):
            result += word1[i]
            i += 1
        while j < len(word2):
            result += word2[j]
            j += 1
        return result
    
# Best solution
# 6ms, beats 95%, using list append and join at end
class Solution:
    def mergeAlternately(self, word1, word2):
        result = [] # Use a list!
        i = 0
        # Use min to zip them together as far as possible
        min_len = min(len(word1), len(word2))
            
        for i in range(min_len):
            result.append(word1[i])
            result.append(word2[i])
            i += 1 # Move to next index

            # Add the remaining tail from whichever string is longer
            # Slicing means it works even if one string
        result.append(word1[i:])
        result.append(word2[i:])
            
        return "".join(result) # Fast to join at end
```

- Explanation and Pattern 
- Anytime you see "Alternatively", "Interleave", "Merge two sorted...", "In-place": Use two pointers to reach O(m+n) time complexity
- For any strings: They are immutable & they can differ in lengths. Use a list [] and append using "".join at the end
- Determine boundary: Are you looping until shortest string ends or longest? 
- How do we handle the tail? 
- Use an index to iterate through the loop: could handle using min of the lengths or a while loop to keep it under the length
- Slicing is fast for handling the longer word

2. Max consecutive ones (EASY)

```python
nums = [1,1,1,1,0,0,0,1,1,0]

class Solution(object):
    def findMaxConsecutiveOnes(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count = 0
        max_count = 0

        for num in nums:
            if num == 1:
                count += 1
                max_count = max(count, max_count)
            else:
                count = 0
        return max_count
```

- Explanation + Pattern 
    - The Running Variable (count): Tracks the current streak.
    - The Global Variable (max_count): Tracks the best result seen so far.
    - The Reset Condition: What breaks the streak? (In this case, a 0).
    - The Final Update: Do I need to check the state one last time after the loop?
- Pattern: Linear Scanning or Greedy
